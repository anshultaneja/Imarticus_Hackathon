```python
m_data['Mean Age'] = m_data['Mean Age'].replace(to_replace=np.nan,value=m_data['Mean Age'].mean())
m_data['Mean Age']
```

```python
m_data.info()
```

```python
m_data.drop(columns=['Severity','Ventilation'],axis = 1,inplace=True)
```

```python
m_data.info()
```

```python
mean = m_data.Hypertension.mean()
mean = round(mean,2)
mean
```

```python
# Replacing Hypertension's null values with mean value
m_data.Hypertension = m_data.Hypertension.replace(to_replace=np.nan,value = mean)
m_data.info()
```

```python
# Replacing null values with their medians
for columns in m_data:
    if columns != 'Mortality':
        m_data[columns] = m_data[columns].replace(to_replace=np.nan,value=m_data[columns].median())
m_data.info()
```

```python
# Replacing Mortality null values with its mean
m_data.Mortality = m_data.Mortality.replace(to_replace=np.nan,value=m_data.Mortality.mean())
m_data.info()
```

```python
# Replacing null values with their medians
for columns in m_data1:
    if columns != 'Mortality':
        m_data1[columns] = m_data1[columns].replace(to_replace=np.nan,value=m_data1[columns].median())
m_data1.info()
```
